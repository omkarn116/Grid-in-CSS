Task 5:


Problem Statement

Explain the difference between justify-items and justify-self using code examples.

Ans:
1. justify-items (container-level)

What it does:
justify-items is applied to the grid container and sets the default horizontal alignment for all grid items inside it.

Common values

- start

- end

- center

- stretch (default)

Example: justify-items

```python
<style>
.grid {
  display: grid;
  grid-template-columns: repeat(3, 120px);
  gap: 10px;
  justify-items: center;
  border: 2px solid black;
}

.item {
  width: 60px;
  height: 60px;
  background: steelblue;
}
</style>

<div class="grid">
  <div class="item">1</div>
  <div class="item">2</div>
  <div class="item">3</div>
</div>
```

2. justify-self (item-level)

What it does:
justify-self is applied to an individual grid item and overrides justify-items for that specific item.

Common values

- start

- end

- center

- stretch

Example: justify-self

```python
<style>
.grid {
  display: grid;
  grid-template-columns: repeat(3, 120px);
  gap: 10px;
  justify-items: start;
  border: 2px solid black;
}

.item {
  width: 60px;
  height: 60px;
  background: tomato;
}

.item.special {
  justify-self: end;
}
</style>

<div class="grid">
  <div class="item">1</div>
  <div class="item special">2</div>
  <div class="item">3</div>
</div>
```
